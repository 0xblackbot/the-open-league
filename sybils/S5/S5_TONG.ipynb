# TONG pump

On Jul 16 @detective_ton TG channel [launched pump of TONG token](https://t.me/detective_ton/101).
It was promoted on the [TON Up platform](https://community.tonup.io/t/the-next-notcoin-level-return-without-tapping-your-screen/1134).
This pump was performed by a numerous wallets, all of them were initialized from one of 4 
special smart contracts, which sent TONs and UP tokens. Also this wallets
used to pump BOOM as well.

To get the list of the sybil wallets Tonalytica query was used:

```sql
select distinct destination_owner, 'TONG' as project, 'Pump' as reason from redoubt.jetton_transfer jt 
where (
source_owner ='EQCYHZkNt3QTShHbVaLIOl-xSmByvisp62MtjgvsBlSIhsJD'
or 
source_owner ='EQBn2UB-oeCKH8NyPnjDLpz0N6F7T-paZzuhKORF53w8b_0V'
or source_owner ='EQAATdYxYmcPEpRqjZ20icYZNj4iPSkRdjziZyEFbhtOKx6m'
or source_owner ='EQCWX9N97QtTUmNxDJT2F5BEagdJ_TOIv_UvvMhOXhvKpz_x'
)
and amount >3000000000000
```
